#Heart Rate by Gender
#grouping by gender to see the average heart rate, blood pressure, and stress

```python
df_sim.groupby('Gender')[['Heart_Rate', 'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic', 'Stress_Level_Biosensor', 'Stress_Level_Self_Report', 'Study_Hours', 'Project_Hours']].mean()
```

The average heart rate for both genders are almost the same (both are 70.1), meaning this simulation assumes physiological uniforminity.
In the simulated environment, there appears to be an inverse relation between the heart rate and biosensor stress when comparing genders. 
    Females (F): Lower average HR (69.7), but Higher Biosensor Stress (5.63).
    Males (M): Higher average HR (70.4), but Lower Biosensor Stress (5.34).
The simulation seems to "view" genders differently, it assumes women's bodies show higher stress through the biosensor compared to men
Study and Project hours are nearly identical, meaning the stress difference isn't coming from a workload gap

My audit will investigate if this is a programed stereotype in the simulation or if the MIT real-world data reflects a similar physciological pattern.

#Using .loc to compare returned rows, the code above assumes "Average Heart Rate", but does not implicitly name it.
#The code below includes logic for "Average Heart Rate" and returns the same rows.

#Using .loc to find females with very high stress and index reset 

```python
High_Stress_F_Sim  = df_sim.loc[
    (df_sim['Gender'] == 'F') &
    (df_sim['Stress_Level_Biosensor'] > 7) &
    (df_sim['Heart_Rate'] > 65) & (df_sim['Heart_Rate'] < 75)
]
High_Stress_F_Sim = High_Stress_F_Sim.reset_index(drop=True)
High_Stress_F_Sim.head()
```

#Using .loc to find males with very high stress and index reset 

```python
High_Stress_M_Sim  = df_sim.loc[
    (df_sim['Gender'] == 'M') &
    (df_sim['Stress_Level_Biosensor'] > 7) &
    (df_sim['Heart_Rate'] > 65 & (df_sim['Heart_Rate'] < 75))
]
High_Stress_M_Sim = High_Stress_M_Sim.reset_index(drop=True)
High_Stress_M_Sim.head()
```
